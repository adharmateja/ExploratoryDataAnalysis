```python
df2 = pd.melt(election_df,id_vars=['State/UT'],value_vars=['Number of constituencies-2009', 'Number of constituencies-2014'])
```

```python
df2.head()
```

```python
const = election_df[['State/UT','Number of constituencies-2009', 'Number of constituencies-2014']]
const.head()
```

```python
plt.figure(figsize=(40,20))
ax=sns.barplot(x="State/UT", y='value',data=df2,hue='variable')
plt.xticks(rotation= 60)
plt.xlabel('States',fontsize=25)
plt.ylabel('Number of Constituences',fontsize=20)
plt.title('State Wise Constituencies in 2009 ande 2014',fontsize=20)
ax.legend(loc='upper right',frameon = True,fontsize=20) 

```

```python
voters = pd.melt(election_df,id_vars=['State/UT'],value_vars=['Number of electors-2009', 'Number of electors-2014'])
voters
```

```python
plt.figure(figsize=(40,20))
ax=sns.barplot(x="State/UT", y='value',data=voters,hue='variable')
plt.xticks(rotation= 60, fontsize=15)
plt.xlabel('States',fontsize=25)
plt.ylabel('Number of Constituences',fontsize=20)
plt.title('State Wise Constituencies in 2009 ande 2014',fontsize=25)
ax.legend(loc='upper right',frameon = True,fontsize=20) 
```

```python
const_2019 = election_df[['State/UT',
       'Number of electors-2009',
       'Number of votes polled-2009'
       , 'Number of valid votes polled-2009',
     ]]
voting_2019 = pd.melt(const_2019,id_vars=['State/UT'],value_vars=['Number of electors-2009',
        'Number of votes polled-2009',
        'Number of valid votes polled-2009'])

voting_2019.head()
```

```python
plt.figure(figsize=(80,20))
ax=sns.barplot(x="State/UT", y='value',data=voting_2019,hue='variable')
plt.xticks(rotation= 60, fontsize=15)
plt.xlabel('States',fontsize=25)
plt.ylabel('Voting Results',fontsize=35)

plt.title('State Wise Voting Results in 2009',fontsize=35)
ax.legend(loc='upper right',frameon = True,fontsize=35) 


```

```python
const_2014 = election_df[['State/UT',
       'Number of electors-2014',
       'Number of votes polled-2014'
       , 'Number of valid votes polled-2014',
     ]]
voting_2014 = pd.melt(const_2014,id_vars=['State/UT'],value_vars=['Number of electors-2014',
        'Number of votes polled-2014',
        'Number of valid votes polled-2014'])

voting_2014.head()
```

```python
plt.figure(figsize=(80,20))
ax=sns.barplot(x="State/UT", y='value',data=voting_2014,hue='variable')
plt.xticks(rotation= 60, fontsize=15)
plt.xlabel('States',fontsize=35)
plt.ylabel('Voting Results',fontsize=35)
plt.title('State Wise Voting Results in 2014',fontsize=35)
ax.legend(loc='upper right',frameon = True,fontsize=35) 

```

```python
percentage = election_df[['State/UT',
       'Percentage of votes polled-2009',
       'Percentage of votes polled-2014',
     ]]
percentage_2009_2014 = pd.melt(percentage,id_vars=['State/UT'],value_vars=[
        'Percentage of votes polled-2009',
       'Percentage of votes polled-2014'])

percentage_2009_2014.head()
```

```python
plt.figure(figsize=(80,35))
ax=sns.barplot(x="State/UT", y='value',data=percentage_2009_2014,hue='variable')
plt.xticks(rotation= 60, fontsize=15)
plt.xlabel('States',fontsize=35)
plt.ylabel('Voting Percentage',fontsize=35)
plt.title('State Wise Voting Percentage in 2009 and  2014',fontsize=35)
ax.legend(loc='upper right',frameon = True,fontsize=35) 

for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2., p.get_height()+0.2, '%d' % int(p.get_height()), 
            fontsize=18, color='blue',weight='bold', ha='center', va='bottom')

plt.show()
```

```python
valid = election_df[['State/UT',
       'Number of valid votes polled-2009',
       'Number of valid votes polled-2014',
     ]]
valid_votes_2009_2014 = pd.melt(valid,id_vars=['State/UT'],value_vars=[
        'Number of valid votes polled-2009',
       'Number of valid votes polled-2014'])

valid_votes_2009_2014.head()
```

```python
plt.figure(figsize=(80,35))
ax=sns.barplot(x="State/UT", y='value',data=valid_votes_2009_2014,hue='variable')
plt.xticks(rotation= 60, fontsize=15)
plt.xlabel('States',fontsize=35)
plt.ylabel('Voting Percentage',fontsize=35)
plt.title('Valid vote polling in 2009 and  2014',fontsize=35)
ax.legend(loc='upper right',frameon = True,fontsize=35) 

```
