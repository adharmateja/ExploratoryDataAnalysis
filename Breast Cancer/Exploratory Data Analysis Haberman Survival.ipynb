### Univariate Analysis 

```python
feature_names = haberman_df.columns
```

```python
feature_names
```

```python
for val in range(len(feature_names[:-1])):
    sns.set_style("whitegrid")
    sns.FacetGrid(haberman_df, hue="status", height=6) \
   .map(sns.distplot, feature_names[val]) \
   .add_legend();
    plt.show();    
```

```python
#CDF 

sns.set_style("whitegrid")

for val in feature_names[:-1]:
    plt.figure(figsize=(10,5))

    counts, bin_edges = np.histogram(haberman_df[val], bins=10, 
                                 density = True)
    pdf = counts/(sum(counts))
    print("PDF: {}".format(pdf))
    print(bin_edges);
    print("BIN: {}".format(bin_edges))

    cdf = np.cumsum(pdf)
    print("CDF: {}".format(cdf))

    plt.plot(bin_edges[1:],pdf);
    plt.plot(bin_edges[1:], cdf)
    plt.xlabel(val)
    plt.show();
    print("\n")
    print("\n")
```

```python
#Box Plots
sns.set_style("whitegrid")


    
fig, axes = plt.subplots(1, 3, figsize=(20, 8))
for idx, val in enumerate(list(haberman_df.columns)[:-1]):
    sns.boxplot( x='status', y=val, data=haberman_df, ax=axes[idx])
plt.show()  
```

```python
#Violin PLot

fig, axes = plt.subplots(1, 3, figsize=(20, 8))
for idx, val in enumerate(list(haberman_df.columns)[:-1]):
    sns.violinplot( x='status', y=val, data=haberman_df, ax=axes[idx])
plt.show() 
```
