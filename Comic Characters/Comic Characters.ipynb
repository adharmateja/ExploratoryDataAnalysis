```python
dc_CSV = pd.read_csv("dc-wikia-data.csv")
marvel_CSV = pd.read_csv("marvel-wikia-data.csv")

dc_CSV.columns = map(str.lower, dc_CSV.columns)
marvel_CSV.columns = map(str.lower, marvel_CSV.columns)

dc_CSV['industry'] = 'DC'
print(len(dc_CSV))
marvel_CSV['industry'] = 'MV'

```

```python
dc_CSV.head()
```

```python
marvel_CSV.head()

marvel_CSV.info()
```

```python
final = pd.concat([dc_CSV,marvel_CSV], sort=False)


final.head()
```

```python

#sex

final.loc[:, 'sex'].replace(['Male Characters','Female Characters','Genderless Characters','Transgender Characters','Genderfluid Characters','Agender Characters'],
                            ['Male','Female','Genderless','Transgender','Genderfluid','Agender'], inplace=True)

#eye
final['eye'].astype('str', inplace=True)
testEye = final.eye.unique()
eye = []
for val in testEye:
    if type(val) is str:
        x = val.split(" ")
        eye.append(x[0])
    else:
        eye.append(val)
        
        
final.loc[:, 'eye'].replace(testEye, eye, inplace = True)

#hair
final['hair'].astype('str',inplace = True)
testHair = final.hair.unique()
hair = []

for val in testHair:
    if type(val) is str:
        x = val.split(" ")
        if len(x)>=3:
            inStr = x[0]+" " +x[1]
            hair.append(inStr)
        elif len(x)==2:
            hair.append(x[0])
        else:
            hair.append(val)
    else:
        hair.append(val)

        
final.loc[:, 'hair'].replace(testHair, hair, inplace = True)

final['align'].unique()

final.loc[:,'align'].replace(['Good Characters', 'Bad Characters', 'Neutral Characters', np.nan,
       'Reformed Criminals'],['Good', 'Bad', 'Neutral', np.nan,
       'Reformed'], inplace=True)



changedYear = final.year.fillna(0).astype(int)

final['year'] = changedYear

final.appearances.isnull().value_counts()


cApp = final.appearances.fillna(0).astype(int)

final['appearances'] = cApp


final.drop(columns=['gsm'],axis=1)

```

```python
final.drop('gsm', axis=1)

final.loc[:,'alive'].replace([np.nan],['Not Defined'],inplace=True)
```

```python
final.alive.unique()
```

```python
final['alive'].astype(str)

final['alive'] = final.loc[:,'alive'].replace(['Living Characters','Deceased Characters','Not Defined'],
                             ['Living','Deceased','None'])


final.head()

del final['gsm']
```

```python

final.loc[:,'first appearance'].replace([np.nan],['None'],inplace=True)
final['first appearance']


oldDate = final['first appearance']
updatedDate = []
for val in oldDate:
    x = val.split(", ")
    if len(x)==2:
        updatedDate.append(x[1])
    else:
        updatedDate.append(val)

        

final['first appearance'] =updatedDate


final.rename(columns={'first appearance': 'month'}, inplace=True)


```
