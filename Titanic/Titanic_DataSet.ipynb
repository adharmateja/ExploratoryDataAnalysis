```python
def age_classification(passenger):
    age, sex = passenger
    
    if age < 16:
        return 'child'
    else:
        return sex
```

```python
titanic_dataset['Person'] = titanic_dataset[['Age','Sex']].apply(age_classification,axis=1)
```

```python
sns.factorplot('Pclass', data =titanic_dataset,kind="count", hue='Person')
```

```python
def change_deck(cabin):
       if type(cabin) == float:
            return cabin
       else:
            return cabin[0]
            
```

```python
titanic_dataset = pd.read_csv("train.csv")
titanic_dataset['Cabin'] = titanic_dataset.Cabin.apply(change_deck)
```

```python
titanic_dataset
```

```python
sns.factorplot('Cabin', data =titanic_dataset,kind="count", palette ='winter_d')
```
